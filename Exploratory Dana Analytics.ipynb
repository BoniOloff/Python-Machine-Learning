```python
# import used packages
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
def get_species (x): 
    if x==0: 
        return 'setosa' 
    elif x==1: 
        return 'versicolor' 
    else: 
        return 'virginica'
```

```python
df['species'] = [get_species(x) for x in iris.target]
```

```python
X = iris.data
y = iris.target
```

```python
pd.plotting.scatter_matrix(df, c=y, s=150, marker='D', figsize=[12,12])
```

```python
sns.set_style('whitegrid')
sns.pairplot(df, hue='species')
```
